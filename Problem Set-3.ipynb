```python
## Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. 
## Label both lines and show the legend.
## Don't forget to label your axes!
```

```python
import numpy as np       #Importing Necessery Libraries
import pandas as pd
import matplotlib.pyplot as plt

us_marriage = pd.read_csv('us-marriages-divorces-1867-2014.csv')
columns = us_marriage.columns.drop(['Marriages','Divorces','Population','Year'])
fig, axes = plt.subplots()
for column in columns:
    axes.plot(us_marriage['Year'], us_marriage[column])       # Chart Plot
    axes.set_title('Marriages & Divorces in US')              # Chart Title
    axes.set_xlabel('Year')                                   # X-axis
    axes.set_ylabel('Divorces')                               # Y-axis
```

```python
# Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.
# Don't forget to label your axes!
```

```python
us_marriage = pd.read_csv('us-marriages-divorces-1867-2014.csv')
s_marriage = us_marriage[(us_marriage.Year == 1900) | (us_marriage.Year == 1950) | (us_marriage.Year == 2000)]
s_marriage = s_marriage.drop(columns = ['Marriages', 'Divorces', 'Population'])
s_marriage = s_marriage.set_index('Year')
s_marriage.plot.bar()
plt.xlabel('Year')                                # X-Axis Label
plt.ylabel('Number of Divorces')                  # Y-Axis Label
plt.title('US Marriages & Divorces per capita')   # Chart Title
```

```python
# Create a horizontal bar chart that compares the deadliest actors in Hollywood.
# Sort the actors by their kill count and label each bar with the corresponding actor's name.
# Don't forget to label your axes!
```

```python
# actor_kill_counts.csv


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


actor_kill_counts = pd.read_csv('actor_kill_counts.csv')
actor_kill_counts.sort_values(by = 'Count', ascending = False, inplace = True)
y = list(actor_kill_counts['Actor'])
x = list(actor_kill_counts['Count'])
plt.barh(y,x)
plt.xlabel('Kill Count')
plt.ylabel('Actors')
plt.title('Hollywood Actor Kill Counts')
for value, index in enumerate(x):
    plt.text(index, value, str(index))
```

```python
## Create a pie chart showing the fraction of all Roman Emperors that were assassinated.
## Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.
```

```python
import pandas as pd

import matplotlib.pyplot as plt


roman_emperor_reigns = pd.read_csv('roman-emperor-reigns.csv')
c = roman_emperor_reigns.groupby('Cause_of_Death').count()[['Emperor']]
c
emperor = pd.Series(c['Emperor'])
cause_death = pd.Series(c.index)
explode = (0.1,0,0,0,0,0,0,0)
fig, ax = plt.subplots()
ax.pie(emperor, explode=explode, labels=cause_death,autopct='%1.1f%%', shadow=False, startangle=90)
ax.axis('equal')
plt.title('Roman Emperors Cause of Death')
plt.show()
```

```python
# Create a scatter plot showing the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009.
# Don't forget to label your axes!
# Color each dot according to its year.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt




arcadeVScs = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')
colors = {2000:'b',2001:'g', 2002:'r',2003:'c',2004:'m', 2005:'y',2006:'k',2007:'w', 2008:'b',2009:'g'}
fig1, ax1 = plt.subplots()
for i in range (len(arcadeVScs['Total Arcade Revenue (billions)'])):
    ax1.scatter(arcadeVScs['Total Arcade Revenue (billions)'][i],arcadeVScs['Computer Science Doctorates Awarded (US)'][i], color = colors[arcadeVScs['Year'][i]])
ax1.set_title('Relationship between total revenue and Number of CS PhDs Awarded in US')
ax1.set_xlabel('Total Arcade Revenue')
```
